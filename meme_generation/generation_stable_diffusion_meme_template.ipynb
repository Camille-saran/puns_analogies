```python
# font
font = cv2.FONT_HERSHEY_SIMPLEX

# fontScale
fontScale = 3
# Blue color in BGR
color = (0, 0, 0)
  
# Line thickness of 2 px
thickness = 5
```

```python
def short(word):
  if len(word)<6:
    return True
```

```python
if short(a_str):
  org_a = (250, 150)
else:
  org_a = (200, 150)
if short(b_str):
  org_b = (1100, 150)
else:
  org_b = (900, 150)
if short(c_str):
  org_c = (250, 1000)
else:
  org_c = (200, 1000)
if short(d_str):
  org_d = (1100, 1000)
else:
  org_d = (900, 1000)

meme = cv2.putText(meme, a_str.upper(), org_a, font,fontScale, color, thickness, 150)
meme = cv2.putText(meme, b_str.upper(), org_b, font,fontScale, color, thickness, 150)
meme = cv2.putText(meme, c_str.upper(), org_c, font,fontScale, color, thickness, 1000)
meme = cv2.putText(meme, d_str.upper(), org_d, font,fontScale, color, thickness, 1000)
```
